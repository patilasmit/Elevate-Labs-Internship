### Rename columns to be lowercase and use underscores for no spaces

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
print(df.columns.tolist())
```

### Data Cleaning and handeling missing values

#### 1. Removed unnecessary columns

```python
df.drop(['id', 'z_costcontact', 'z_revenue','complain', 'acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3', 'acceptedcmp4', 'acceptedcmp5'], axis=1, inplace=True)
```

#### 2.Handle Missing Values

```python
df.isnull().sum() 
```

##### Found only 24 missing values in the Income column.

```python
df['income'] = df['income'].fillna(df['income'].median())
```

##### Missing income values filled using median (safer than mean due to outliers)

```python
df['income'].isnull().sum()
```

#### 3. Checking for Duplicates & Drop 

```python
duplicates_count = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates_count}")
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

##### No Duplicate Records found in dataset

#### 4. Text Normalization: Converted Dt_Customer to datetime.

##### 1.Filter columns that have text (object) data type

```python
text_columns = df.select_dtypes(include='object').columns
print("Text columns in the DataFrame:")
print(text_columns)
```

```python
text_columns = ['education', 'marital_status']
for col in text_columns:
    df[col] = df[col].str.lower().str.strip()
print("Categorical text fields normalized")
```

##### 2. Fix inconsistent text cases

```python
df['education'] = df['education'].str.title()
df['marital_status'] = df['marital_status'].str.title()
```

 #### 5. Date Formatting

```python
df.rename(columns={'dt_customer': 'date'}, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'],format="%d-%m-%Y",errors='coerce')
print("Date Formatted Correctly")
```

```python
df.head()
```

#### 6.Clean inconsistent marital status values
##### Cleaning inconsistent category names like 'YOLO', 'Absurd', 'Alone' ➝ maybe grouped under 'Single' or 'Other'
##### Ensure consistency in naming like 'Married', 'Together', 'Divorced', etc.

```python
df['marital_status'] = df['marital_status'].replace({
    'YOLO': 'Single',
    'Absurd': 'Single',
    'Alone': 'Single',
    'Married': 'Married',
    'Together': 'Married',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Widowed': 'Widowed',
    'Single': 'Single'
})
```

### Adding Some Additional Columns(Feature Engineering)

##### 1. Customer Age

```python
df['customer_age'] = 2025 - df['year_birth']
```

```python
print(df['customer_age'].describe())
```

```python
df = df[(df['customer_age'] <= 100) & (df['customer_age'] >= 18)]
```

##### 2. Age Group 

```python
df['age_group'] = pd.cut(df['customer_age'],bins=[0,30,45,60,80,100],labels=['<30','30-45','45-60','60-80','80+'])
```

```python
df.head()
```

##### 3. Day as Customer

```python
reference_date = pd.Timestamp("2025-04-21")
df['day_as_customer'] = (reference_date - df['date']).dt.days
```

```python
df.head()
```

##### 4. Customer Loyalty Tier

```python
df['loyality_tier'] = pd.cut(
    df['day_as_customer'],
    bins=[0, 365, 1095, 1825, 5000],
    labels=['new', 'regular', 'loyal', 'very loyal']
)

```

```python
df['loyality_tier'] = df['loyality_tier'].cat.add_categories(['new', 'regular', 'loyal', 'very loyal'])
```

##### 5. Is Parents or not


```python
df['is_parent'] = (df['kidhome'] > 0).astype(int)
```

##### 6.household_size

```python
df['household_size'] = df['kidhome'] + df['teenhome'] + 1
```

##### 7. All product category spendings 

```python
df["Total_Spending"] = df[[
    'mntwines', 'mntfruits', 'mntmeatproducts',
    'mntfishproducts', 'mntsweetproducts', 'mntgoldprods'
]].sum(axis=1)
```
